```python
import tweepy
import csv
import os
from os import path
import emoji
from nltk.tokenize import word_tokenize
import re, string , time
from datetime import datetime
from nltk.corpus import stopwords
import time

#####function that writes scrappd data in rows in csv####
def insertrow(findword,tweettext,csvWriter):  
    tweettext = emoji.demojize(tweettext)
    csvWriter.writerow([findword, tweettext.encode('ascii', errors='ignore')])

    
#####function that create csv and scrape data from twitter####
def createcsv(api):
    filename="dataset(raw).csv"
    print("Filename ("+filename +")")

    if not path.exists(filename):
        ####Open/Create a file to append data###########
        csvFile = open(filename ,'a',newline="")
        #####Use csv Writer#################
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow(["Emotion","Content"])
        print('Create new csv file')
    else:
        ####Open/Create a file to append data###########
        csvFile = open(filename ,'a',newline="")
        #####Use csv Writer#################
        csvWriter = csv.writer(csvFile)
        print('Found previous csv file')
    
    print("Program has started scraping")
    #####to record start time############
    starttime=time.time()
    searchword="#angry OR #anger OR #furious OR #mad OR #enraged OR #happy OR #joy OR #delighted OR #disappointed OR #upset OR #frown OR #unsatisfied OR #complain OR üò° OR üò† OR ü§¨ OR üëø OR üí¢ OR ü•≥ OR üòÇ OR üòÄ OR üòÅ OR ü§£ OR üòû OR ‚òπÔ∏è"
    count=200
    numberoftweets=0
    for tweet in tweepy.Cursor(api.search,q=searchword,count=count,
                           lang="en",
                           since="2007-01-01").items():
        
        
    
        #####To avoid ConnectionResetError: [WinError 10054] An existing connection was forcibly closed by the remote host####
        #####15,000 tweets/15 mins for app.####
        numberoftweets+=1
        if numberoftweets==15000:
            runtime=time.time()-starttime
            print("Program has already scraped 15000 tweets in " + str(int(runtime/60)) +" min "+ str(int(runtime%60))+"s" )
            if(runtime)<900:
                sleeptime=900-runtime
                print("Program is going to sleep for " + str(int(sleeptime/60)) +" min "+ str(int(sleeptime%60))+"s" )
                time.sleep(sleeptime)
            print("program has been restarted")
            starttime=time.time()
            numberoftweets=0
    
    
        if("#angry" in tweet.text or "#anger" in tweet.text  or "#furious" in tweet.text  or "#mad" in tweet.text  or "#enraged" in tweet.text or  "üò°" in tweet.text or "üò†" in tweet.text  or "ü§¨" in tweet.text  or "üëø" in tweet.text  or "üí¢" in tweet.text):
            findword="angry"
            insertrow(findword, tweet.text,csvWriter)
#         if ("#happy" in tweet.text or "#joy" in  tweet.text or "#delighted" in  tweet.text or "ü•≥" in tweet.text or "üòÇ" in  tweet.text or "üòÄ" in  tweet.text or "üòÅ" in  tweet.text or "ü§£" in  tweet.text):
#             findword="happy"
#             insertrow(findword, tweet.text,csvWriter)
        if ("disappointed" in  tweet.text or "#upset" in  tweet.text or "#frown" in  tweet.text or "#unsatisfied" in  tweet.text or "#complain" in  tweet.text or "üòû" in  tweet.text or "‚òπÔ∏è" in  tweet.text):
            findword="disappointed"
            insertrow(findword, tweet.text,csvWriter)
            
    csvFile.close()

def main():
####Credientials{Put your keys and token between ''}##############################    
    consumer_key = ''
    consumer_secret = ''
    access_token = ''
    access_token_secret = ''

    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth,wait_on_rate_limit=True)
    print('Credientials successful')
    createcsv(api)


##### Standard boilerplate to call the main() function to begin
##### the program.
if __name__== "__main__":
    main()
```
